```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([1, 2, 3, 4, 5])
b = a ** 2
plt.plot(a, b)

plt.title("kvadratik o'sish")
plt.xlabel("X o'qi")
plt.ylabel("Y o'qi")

plt.show
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color = "green", linestyle = "--")
plt.title("sinx")
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(10, 4)) # Rasm o'lchami

# Birinchi grafik (1-qator, 2-ustun, 1-chi rasm)
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Sinus")

# Ikkinchi grafik (1-qator, 2-ustun, 2-chi rasm)
plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x), color='red')
plt.title("Kosinus")

plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.sin(x)

# Birinchi grafik (1-qator, 2-ustun, 1-chi rasm)
plt.subplot(1, 2, 1)
plt.plot(x, y, color = "green", linestyle = "dotted", marker = "*", markersize = 8, markerfacecolor = "yellow", markeredgecolor = "red")
plt.title("Kam nuqtali grafik")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.sin(x)

plt.plot(x, y, color='blue', linestyle='--', linewidth=2, 
         marker='*', markersize=15, markerfacecolor='yellow', markeredgecolor='red')

plt.title("Yulduzchali Markerlar")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 1000)
y = x ** 2 - 4 * x + 4
plt.plot(x, y, color = "blue", label = "f(x)")


###################################
ax = plt.gca() # Hozirgi o'qlarni (axis) qo'lga olamiz

# O'ng va yuqori chiziqlarni ko'rinmas qilamiz
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Pastki o'qni (X) Y=0 nuqtasiga ko'chiramiz
ax.spines['bottom'].set_position(('data', 0))

# Chap o'qni (Y) X=0 nuqtasiga ko'chiramiz
ax.spines['left'].set_position(('data', 0))
##################################
plt.title("f(x) = x^2 - 4x + 4")
plt.xlabel("X array")
plt.ylabel("Y arrray")

plt.grid(True)
plt.legend()
plt.show()
```
