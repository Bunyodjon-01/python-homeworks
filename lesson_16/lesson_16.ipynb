```python
import pyodbc

# Ulanish sozlamalari
server = r'BUNYODJONS\SQLEXPRESS' # Masalan: localhost yoki 192.168.1.1
database = 'exam_preperation'

# Connection string (Ulanish qatori)
# Windows Authentication (parolsiz) bo'lsa:
conn_str = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;'

# SQL Server Authentication (login/parol bilan) bo'lsa:
# conn_str = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}'

try:
    # Ulanishni amalga oshirish
    conn = pyodbc.connect(conn_str)
    cursor = conn.cursor()
    print("Muvaffaqiyatli ulandi!")

    # Ma'lumotlarni o'qish (misol)
    cursor.execute('''
                   SELECT id 
                   FROM dbo.oscar_nominees 
                   where category = 'Best Actor'
                   ''')
    for row in cursor:
        print(row)

    # Ulanishni yopish
    conn.close()

except Exception as e:
    print(f"Xatolik yuz berdi: {e}")
```
