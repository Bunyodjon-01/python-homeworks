```python
import sqlite3

connection = sqlite3.connect("roster.db")
create_table = '''
    create table if not exists "Roster" (
        name nvarchar(100),
        species nvarchar(100),
        age int
        );
'''
values = '''
    insert into Roster
    values
        ("Benjamin Sisko", "Human", 40),
        ("Jadzia Dax", "Trill", 300),
        ("Kira Nerys", "Bajoran", 29);
'''
cursor = connection.cursor()
cursor.execute(create_table)
cursor.execute(values)
connection.commit()
connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("roster.db")

query = '''
    UPDATE Roster
    SET name = 'Ezri Dax'
    WHERE name = 'Jadzia Dax';
'''
cursor = connection.cursor()
cursor.execute(query)
connection.commit()
connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("roster.db")

query = '''
    select name, age from roster
    where species = "Bajoran";
'''
cursor = connection.cursor()
cursor.execute(query)
results = cursor.fetchall()
for i in results:
    print(i)
connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("roster.db")

query = '''
    DELETE from roster
    where age > 100;
'''
cursor = connection.cursor()
cursor.execute(query)
connection.commit()
connection.close()
```

```python
import sqlite3

connection = sqlite3.connect("roster.db")

query = '''
    select * from roster
    order by age desc;
'''
cursor = connection.cursor()
cursor.execute(query)
results = cursor.fetchall()
for i in results:
    print(i)
connection.close()
```
