```python
def format_txt(txt):
    vowels = "aeiouAEIOU"
    result_list = []
    idx = 0
    
    while idx < len(txt):
        # Standart holatda 3 ta harf olamiz
        step = 3
        
        # 1. Tekshiramiz: Agar 3-belgi mavjud bo'lsa
        if idx + 2 < len(txt):
            char_3 = txt[idx + 2]
            
            # 2. Agar 3-belgi unli bo'lsa, chiziqchani bitta suramiz (ya'ni 4 ta harf olamiz)
            if char_3 in vowels:
                step = 4
            
            # 3. "Orqasidan ostki chiziqcha qo'yilgan harf" (o' va g' uchun)
            # Agar 3-belgi 'o' yoki 'g' bo'lsa va undan keyin ' kelsa
            elif idx + 3 < len(txt) and txt[idx + 3] == "'":
                step = 4
        
        # Bo'lakni kesib olamiz
        segment = txt[idx : idx + step]
        result_list.append(segment)
        
        # Eng muhimi: Ko'rsatkichni qadam miqdoriga suramiz (Infinite loop bo'lmasligi uchun)
        idx += step
        
    # Bo'laklarni '_' bilan birlashtiramiz
    return "_".join(result_list)

# Tekshirish:
print(format_txt("hello"))                     # hel_lo
print(format_txt("assalom"))                   # ass_alom
print(format_txt("abcabcdabcdeabcdefabcdefg")) # abc_abc_dab_cdea_bcd_efab_cdef_g
```

```python
def format_txt(txt):
    vowels = "aeiouAEIOU"
    result = ""
    count = 0
    i = 0
    
    while i < len(txt):
        result += txt[i]
        count += 1
        
        # Agar kamida 3 ta belgi to'plangan bo'lsa
        if count >= 3:
            # SHART: 
            # 1. Hozirgi harf unli bo'lmasligi kerak
            # 2. Keyingi belgi tutuq belgisi (') bo'lmasligi kerak
            # 3. Bu oxirgi belgi bo'lmasligi kerak
            
            is_current_vowel = txt[i] in vowels
            has_apostrophe_next = (i + 1 < len(txt) and txt[i+1] == "'")
            
            if not is_current_vowel and not has_apostrophe_next:
                if i < len(txt) - 1: # Oxirgi belgi bo'lmasa underscore qo'sh
                    result += "_"
                    count = 0 # Countni nolga tushiramiz
        
        i += 1
    
    # Ba'zida oxirida _ qolib ketishi mumkin, shuni tekshiramiz
    return result.rstrip("_")

# Tekshiruv
print(format_txt("hello"))                     # hel_lo
print(format_txt("assalom"))                   # ass_alom
print(format_txt("abcabcdabcdeabcdefabcdefg"))
```

```python
import random
while True:
    num = random.randint(1, 100)
    i = 1
    while i <= 10:
        a = int(input())
        if a == num:
            print("You guessed it right!")
            break
        elif a > num:
            print("Too high!")
        else:
            print("Too low!")
        i += 1
    else :
        print("you lost...")
    reply = input("You lost. Want to play again?")

    if reply not in ['Y', 'YES', 'y', 'yes', 'ok' ]:
        break
```

```python
while True:
    password = input("Enter password: ")

    if len(password) < 8:
        print("Password is too short.")
    elif not any(ch.isupper() for ch in password):
        print("Password must contain an uppercase letter.")
    else:
        print("Password is strong.")
        break
```
